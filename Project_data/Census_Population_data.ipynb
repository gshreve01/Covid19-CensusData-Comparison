```python
import matplotlib.pyplot as plt 
import pandas as pd
import numpy as np
import requests
import gmaps
import os
import csv
import json 
```

```python
url="https://api.census.gov/data/2019/pep/population?get=COUNTY,DENSITY,POP,NAME,STATE&in=state:37&for=county:*&key=972cf3f08f68b6e55893b0ff405867f65ac840eb"
response = requests.get(url).json()
#print(json.dumps(response, indent=4, sort_keys=True))
df = pd.DataFrame(response)
df
```

```python
df1 =df.rename(columns = {0:'County',1:'Density',2:'Population',3:'County_State',4:'State',
                    5:'State Code',6:'County Code'})
df2 = df1.drop(index =[0])
df2
```

```python

new = df2["County_State"].str.split(",", n = 1, expand = True) 
  
# making separate first name column from new data frame 
df2["County_Name"]= new[0] 
  
# making separate last name column from new data frame 
df2["State_Name"]= new[1]
df2
```

```python
df2.to_csv("Census_Population_data.csv",index = False)
```
